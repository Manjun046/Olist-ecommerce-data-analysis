```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
    
]

for col in date_cols:
    df_orders[col] = pd.to_datetime(df_orders[col], errors='coerce')
```

```python
df_orders[df_orders['order_delivered_customer_date'].isnull()]['order_status'].value_counts()
```

```python
delivered_orders = df_orders[
    df_orders['order_status'] == 'delivered'
].copy()


delivered_orders.shape
```

```python
delivered_orders.isnull().sum()
```

```python
delivered_orders.dtypes
```

```python
delivered_orders = delivered_orders.dropna(
    subset=[
        'order_approved_at',
        'order_delivered_carrier_date',
        'order_delivered_customer_date',
        'order_estimated_delivery_date',
    ]
)
delivered_orders.shape
```

```python
delivered_orders.isnull().sum()
```

```python
delivered_orders['delivery_days'] = (
    delivered_orders['order_delivered_customer_date']
    - delivered_orders['order_purchase_timestamp']).dt.days
```

```python
delivered_orders.dtypes
```

```python
delivered_orders['delivery_delay_days'] = (
    delivered_orders['order_delivered_customer_date']- delivered_orders['order_estimated_delivery_date']
).dt.days
```

```python
delivered_orders.to_csv('Cleaned_orders',index= False)
```

```python
import csv

# Convert to string and remove problematic quotes
for col in ['review_comment_title', 'review_comment_message']:
    df_reviews[col] = (
        df_reviews[col]
        .fillna('No comment')
        .astype(str)
        .str.replace('"', '', regex=False)
        .str.replace('\n', ' ', regex=False)
        .str.replace('\r', ' ', regex=False)
    )

df_reviews.to_csv(
    'Cleaned_reviews_final.csv',
    index=False,
    quoting=csv.QUOTE_ALL
)
```
